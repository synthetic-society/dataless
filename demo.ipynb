```python
from dataless.extrapolate import PYPExtrapolation

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d = pd.DataFrame({'n': [1, 10, 100], 'κ': [1, 0.99, 0.95]})
d
```

```python
model = PYPExtrapolation(d)
model.train()
model.test(np.array([1, 10, 100, 1000, 10000]))
```

```python
fig, ax = plt.subplots()

n = np.exp(np.log(10) * np.linspace(0, 10, 100))
ax.plot(d.n, d.κ, 'o', label='Original data')
ax.plot(n, model.test(n), label='Extrapolation')
ax.set(xscale='log', xlabel='Population (n)', ylabel='Identification accuracy (κ)', ylim=(0, 1))
ax.legend();
```
