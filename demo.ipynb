```python
import matplotlib.pyplot as plt
import numpy as np
from dataless import PYPExtrapolation

%matplotlib inline
```

```python
n = [1, 10, 100]
correctness = [1, 0.99, 0.90]
model = PYPExtrapolation(n, correctness=correctness)
model.predict(np.array([1, 10, 100, 1000, 10000]))
```

```python
fig, ax = plt.subplots()

n_scale = np.exp(np.log(10) * np.linspace(0, 10, 100))
correctness_scale = model.predict(n_scale)
ax.plot(n, correctness, "o", label="Original data")
ax.plot(n_scale, correctness_scale, label="Extrapolation")
ax.set(xscale="log", xlabel="Population (n)", ylabel="Identification accuracy (Îº)", ylim=(0, 1))
ax.legend();
```
